```python
columns = list(telcom.columns)
```

```python
def missing_rows(telcom, columns):
    missing_rows = {}
    for column in columns:
        missing_rows[column] = telcom[column].isna().sum()
    return missing_rows
```

```python
missing_row = missing_rows(telcom, columns)
total_rows = telcom.shape[0]
for column,count in missing_row.items():
    print(f'{column}: {count} missing values, percentage: {count/total_rows:.2%}')
```

```python
def unique_values(telcom, columns):
    return {column:telcom[column].unique().tolist() for column in columns}
```

```python
u_values = unique_values(telcom, columns)
def detect_empty_string(unique_values):
    empty_string_columns = []
    invalid_tokens = {'', 'NA', 'N/A', 'na', 'n/a', 'none'}
    for column, values in unique_values.items():
        normalized = {str(value).strip().lower() for value in values}
        if normalized & invalid_tokens:
            empty_string_columns.append(column)
    return empty_string_columns        
```

```python
print(detect_empty_string(u_values))
```

```python
total_charges_unique = telcom['Total Charges'].unique().tolist()
def detect_invalid_string(total_charges_unique):
    invalid_tokens = {'', 'NA', 'N/A', 'na', 'n/a', 'none'}
    normalized = {str(value).strip().lower() for value in total_charges_unique}
    detected_tockens = normalized & invalid_tokens
    return detected_tockens
```

```python
print(detect_invalid_string(total_charges_unique))
```
